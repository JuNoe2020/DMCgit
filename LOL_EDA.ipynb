### 데이터 분포 형태 파악

```python
df.columns
```

##### gameId는 분석에 필요없으므로 drop

##### blueKills, blueDeaths, blueWardsPlaced, blueWardsDestroyed,  blueAssists, blueTowersDestroyed, blueTotalGold, blueAvgLevel, blueTotalExperience, blueTotalMinionsKilled, blueTotalJungleMinionsKilled 변수들은 절대수치보다 상대팀보다 얼마나 높은지가 중요하므로 redKills, redDeaths, redWardsPlaced, redWardsDestroyed,  redAssists, redTowersDestroyed, redTotalGold, redAvgLevel, redTotalExperience, redTotalMinionsKilled, redTotalJungleMinionsKilled 의 비율로 나타낸다.

##### redGoldDiff, redExperienceDiff는 blueGoldDiff, blueExperienceDiff와 거의 유사한 변수들로 blue팀이 특정값일 경우 red팀이 -(특정값)이므로 불필요하기 때문에 redGoldDiff와 redExperienceDiff는 drop한다.

##### redEliteMonsters, redDragons, redHeralds는 blueEliteMonsters, blueDragons, blueHeralds와 중복되는 더미변수이므로 drop

##### blueTotalGold은 blueTotalMinionsKilled, blueKills와 강한 상관관계가 예측되지만 상관분석 후 판단하기로 한다. (vice versa for red)

```python
# gameId drop 
df = df.drop(['gameId'], axis=1)
df
```

```python
# blueKills, blueDeaths, blueWardsPlaced, blueWardsDestroyed, blueAssists, blueTowersDestroyed, blueTotalGold, blueAvgLevel, 
# blueTotalExperience, blueTotalMinionsKilled, blueTotalJungleMinionsKilled 변수들은 절대수치보다 상대팀보다 얼마나 높은지가 
# 중요하므로 redKills, redDeaths, redWardsPlaced, redWardsDestroyed, redAssists, redTowersDestroyed, redTotalGold, redAvgLevel, 
# redTotalExperience, redTotalMinionsKilled, redTotalJungleMinionsKilled 의 비율로 나타낸다.

df['blueKillsRatio'] = df['blueKills']/df['redKills'] 
df['blueDeathsRatio'] = df['blueDeaths']/df['redDeaths']
df['blueWardsDestroyedRatio']= df['blueWardsDestroyed']/df['redWardsDestroyed']
df['blueAssistsRatio'] = df['blueAssists']/df['redAssists']
df['blueTowersDestroyedRatio'] = df['blueTowersDestroyed']/df['RedTowersDestroyed']
df['blueTotalGoldRatio'] = df['blueTotalGold']/df['redTotalGold']
df['blueAvgLevelRatio'] = df['blueAvgLevel']/df['redAvgLevel']
df['blueTotalExperienceRatio'] = df['blueTotalExperience']/df['redTotalExperience']
df['blueTotalMinionsKilledRatio'] = df['blueTotalMinionsKilled']/df['redTotalMinionsKilled']
df['blueTotalJungleMinionsKilledRatio'] = df['blueTotalJungleMinionsKilled']/df['redTotalJungleMinionsKilled']

df.drop(['blueKills', 'blueDeaths', 'blueWardsPlaced', 'blueWardsDestroyed', 'blueAssists', 'blueTowersDestroyed',\
         'blueTotalGold', blueAvgLevel, 
# blueTotalExperience, blueTotalMinionsKilled, blueTotalJungleMinionsKilled ])
```
